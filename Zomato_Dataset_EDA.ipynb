# Zomato maximum transactions are from India followed by USA then UK

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'}) #first grouping some selected columns then finding the size for each category then taking index along with the size column name and then renaming the size column. 
```

```python
ratings
```

*  When the average rating is **4.5 to 4.9** it indicates that the food is **excellent**.
*  When the average rating is **4.0 to 4.4** it indicates that the food is **very good**.
*  When the average rating is **3.5 to 3.9** it indicates that the food is **good**.
*  When the average rating is **2.5 to 3.4** it indicates that the food is **average**.
*  When the average rating is **1.8 to 2.4** it indicates that the food is **poor**.

```python
ratings.head()
```

```python
#plotting the graph in the same color as of the rating colors
import matplotlib
matplotlib.rcParams['figure.figsize']= (10,5)
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color",data=ratings, palette =['grey','red','orange','yellow','green','green'])
plt.xticks(rotation=90)
```

*  Not rated count is very high compared to any other single rating.
*  Maximum number of rating are between 2.8 to 3.9





```python
  #counting the frequency of the colors present in the ratings df
  sns.countplot(x="Rating color", data=ratings,palette =['grey','red','orange','yellow','green','green'] )  
```
